```python
import pandas as pd
import numpy as np
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import GridSearchCV, cross_validate
from sklearn.preprocessing import LabelEncoder, StandardScaler
from sklearn.metrics import make_scorer, precision_score, recall_score, f1_score, accuracy_score
from imblearn.over_sampling import SMOTE
import joblib
import matplotlib.pyplot as plt
import seaborn as sns

# Carregar os dados
df = pd.read_csv('/content/drive/MyDrive/Aula 2/dados/Regensburg Pediatric Appendicitis.csv', sep=',')

# Remover coluna de índice não nomeada, se existir
df = df.drop(columns=['Unnamed: 0'], errors='ignore')

# Tratar valores ausentes - moda para categóricas e mediana para numéricas
for col in df.columns:
    if df[col].dtype == 'object':
        df[col] = df[col].fillna(df[col].mode()[0])
    else:
        df[col] = df[col].fillna(df[col].median())

# Codificar automaticamente todas as colunas categóricas
label_encoders = {}
for col in df.select_dtypes(include=['object']).columns:
    le = LabelEncoder()
    df[col] = le.fit_transform(df[col].astype(str))
    label_encoders[col] = le

# Verificar se todas as colunas estão numéricas
non_numeric_cols = df.select_dtypes(include=['object']).columns
if len(non_numeric_cols) > 0:
    print("\n⚠️ Ainda existem colunas não numéricas:")
    print(non_numeric_cols)
else:
    print("\n✅ Todas as colunas estão numéricas.")

# Preparar dados (features)
X = df.drop(columns=['Diagnosis', 'Severity', 'Management'])

# Normalização dos dados
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Função para especificidade
def specificity_score(y_true, y_pred):
    cm = pd.crosstab(y_true, y_pred)
    TN = cm.loc[0, 0] if (0,0) in cm else 0
    FP = cm.loc[0, 1] if (0,1) in cm else 0
    return TN / (TN + FP) if (TN + FP) > 0 else 0

specificity = make_scorer(specificity_score)

# Balanceamento usando SMOTE (para diagnóstico)
y_diagnosis = df['Diagnosis']
smote = SMOTE(random_state=42)
X_bal, y_bal = smote.fit_resample(X_scaled, y_diagnosis)

# Modelo e hiperparâmetros
rf = RandomForestClassifier(random_state=42)
param_grid = {
    'n_estimators': [50, 100, 200],
    'max_depth': [None, 10, 20],
    'min_samples_split': [2, 5],
    'min_samples_leaf': [1, 2]
}

# Métricas para avaliação
scoring = {
    'accuracy': 'accuracy',
    'precision': 'precision',
    'recall': 'recall',
    'f1': 'f1',
    'specificity': specificity
}

# GridSearchCV com cross-validation para achar melhores parâmetros
grid_search = GridSearchCV(rf, param_grid, scoring=scoring, refit='accuracy', cv=5)
grid_search.fit(X_bal, y_bal)

print("\nMelhores parâmetros encontrados para Diagnóstico:", grid_search.best_params_)
print(f"Melhor acurácia média (CV): {grid_search.best_score_:.4f}")

# Salvar modelo e scaler
joblib.dump(grid_search.best_estimator_, 'model_rf_diagnosis.pkl')
joblib.dump(scaler, 'scaler.pkl')
joblib.dump(label_encoders, 'label_encoders.pkl')

# Avaliar com cross_validate usando melhores parâmetros
best_rf = grid_search.best_estimator_
cv_results = cross_validate(best_rf, X_bal, y_bal, cv=5, scoring=scoring)

df_cv_results = pd.DataFrame(cv_results).mean()
print("\nMédias das métricas na validação cruzada:")
print(df_cv_results)

# Importância das features
feature_importances = pd.DataFrame(best_rf.feature_importances_, index=X.columns, columns=['importance']).sort_values('importance', ascending=False)
print("\nTop 10 características importantes para Diagnóstico:")
print(feature_importances.head(10))

# Visualizar importância das features
plt.figure(figsize=(10,6))
sns.barplot(x=feature_importances.head(10)['importance'], y=feature_importances.head(10).index)
plt.title('Top 10 Características para Diagnóstico')
plt.xlabel('Importância')
plt.ylabel('Característica')
plt.show()
```
